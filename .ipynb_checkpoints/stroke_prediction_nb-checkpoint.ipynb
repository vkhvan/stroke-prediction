```python
import pandas as pd
import plotly as pl
import seaborn as sb
```

```python
df = pd.read_csv("./data/healthcare-dataset-stroke-data.csv")
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['id'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ['bmi'], inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df_s = df.groupby(['stroke'])[df.columns[0]].count().reset_index(name='cases')
print(df_s)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = df_s['cases']
labels = df_s['stroke']

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
df.head()
```

```python
gender_dummies = pd.get_dummies(df.gender, prefix='Gender').iloc[:, 0:]
df = pd.concat([df, gender_dummies], axis=1)
df.head()
```

```python
ever_married_dummies = pd.get_dummies(df.ever_married, prefix='Ever_Married').iloc[:, 0:]
df = pd.concat([df, ever_married_dummies], axis=1)
df.head()
```

```python
work_type_dummies = pd.get_dummies(df.work_type, prefix='Work_Type').iloc[:, 0:]
df = pd.concat([df, work_type_dummies], axis=1)
df.head()
```

```python
residence_type_dummies = pd.get_dummies(df.Residence_type, prefix='Residence_Type').iloc[:, 0:]
df = pd.concat([df, residence_type_dummies], axis=1)
df.head()
```

```python
smoking_status_dummies = pd.get_dummies(df.smoking_status, prefix='Smoking_Status').iloc[:, 0:]
df = pd.concat([df, smoking_status_dummies], axis=1)
df.head()
```

```python
df_s = df.drop(columns=['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'])
```

```python
df_s.head()
```

```python
import pandas as pd
 
# use corr() method on dataframe to
# make correlation matrix
matrix = df_s.corr()
 
# print correlation matrix
print("Correlation Matrix is : ")
print(matrix)
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Compute the correlation matrix
corr = df_s.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
